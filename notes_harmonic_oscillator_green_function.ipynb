### harmonic oscillator

The equation for the harmonic oscillator is:

\begin{equation}
F/m = \ddot{y} = - \frac{k}{m} y - 2 b \dot{y} + f(t)
\end{equation}

with $\omega_0^2 = \frac{k}{m}$ follows

\begin{equation}
\ddot{y} + 2 b \dot{y} + \omega_0^2 y = f(t) \\
\underbrace{\Big( \frac{d^2}{dt^2} + 2 b \frac{d}{dt} + \omega_0^2 \Big)}_{L} y = f(t)
\end{equation}

### green functions

It can be solved by a green function. A green function $G(t,t_0)$ is defined as the solution to the above equation wich returns a delta distribution $L G(t,t_0) = f(t)=\delta(t-t_0)$. It is the impulse response of the particular system ( the harmonic oscillator in this case). Since any function can be build by adding Delta-peaks

\begin{align}
f(t) = \int \delta(t-t_0) f(t_0) dt_0
\end{align}

the general solution for the differential equation can be computed as:

\begin{align}
y_p(t) &= L f(t) \\
&= L \int \delta(t-t_0) f(t_0) dt_0 \\
&= \int G(t,t_0) f(t_0) dt_0 \\
\end{align}

### actual green function

The green function for the harmonic oscillator is $G(t,t_0)=0$ for $t<t_0$ and :

\begin{align}
G(t,t_0) = \frac{1}{r_1-r_2} \big[ e^{r_1 (t-t_0)} + e^{r_2 (t-t_0)} \big]
\end{align}

for $t>t_0$.

### homogeneous solution

additional one needs the homogeneous solution:

\begin{equation}
\ddot{y} + 2 b \dot{y} + \omega_0^2 y = 0 \\
\end{equation}

with the guess

\begin{equation}
y_h(t) = C_1 e^{r_1 t} + C_2 e^{r_2 t}
\end{equation}

this equation simplifies to

\begin{equation}
r_i^2 + 2 b r_i + \omega_0^2 = 0 \\
\end{equation}

and has following solution

\begin{equation}
r_i = b \pm \sqrt{b^2 - \omega_0^2} \\
\end{equation}

### general solution


\begin{align}
y(t) &= y_h(t) + y_p(t) \\
&= C_1 e^{r_1 t} + C_2 e^{r_2 t} + \int \frac{1}{r_1-r_2} \big[ e^{r_1 (t-t_0)} + e^{r_2 (t-t_0)} \big] f(t_0) dt_0 \\
\end{align}

with $r_i = b \pm \sqrt{b^2 - \omega_0^2}$
