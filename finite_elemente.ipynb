## Finite Elemente Methode

### Einheiten

Volumen $V = m^3$<br>
Temperatur $T = K$<br>
Dichte $\rho = \frac{kg}{m^3}$<br>
Wärmestromdichte $Q = W = \frac{J}{s}$<br>
Wärmestromdichte $q = \frac{W}{m^2} = \frac{J}{m^2 s}$<br>
Wärmequelle-Dichte $\varphi = \frac{W}{m^3}$<br>
Wärmekapazität $c = \frac{J}{kg K}$<br>
Wärmeleitfähigkeit $k \hspace{0.2em} \textrm{oder} \hspace{0.2em} \lambda = \frac{W}{m K}$<br>
Temperaturleitfähigkeit $\alpha = \frac{m^2}{s}$<br>
Wärmequellenterm $\phi = \frac{J}{m^3 s}$<br>

### Herleitung der Wärmeleitung über Differenzen

Wärmeleitung in Isolatoren beruht auf Gitterschwingungen (Phononen) und in Metallen zusätzlich auf freien Elektronen. In Flüssigkeiten und Gasen passiert Wärmeleitung außerdem durch Vermischung (Konvektion). Ein weiterer Mechanismus der Wärmeübertragung ist Wärmestrahlung. Die gewöhnliche Wärmeleitungsgleichung beschreibt durch das Fouriergesetzt Wärmetransport in einem Festkörper, so dass Konvetion und Wärmestrahlung nicht berücksichtigt werden.
<br>

Änderung der Temperatur = Wärmefluss durch die Flächen eines Quarders + innere Wärmequelle. <br>
\begin{equation}
\Delta (c \rho V T) = (q_{x} - q_{x+\Delta x}) \Delta y \Delta z \Delta t + (q_{y} - q_{y+\Delta y}) \Delta x \Delta z \Delta t + (q_{z} - q_{z+\Delta z}) \Delta x \Delta y \Delta t + \varphi(x,y,z,t)  \Delta x \Delta y \Delta z \Delta t
\end{equation}
<br>

Umformung und $\Delta x,\Delta y,\Delta z,\Delta t \rightarrow 0$: <br>
\begin{equation}
\frac{d}{d t} (c \rho T) = - \nabla \vec{q} + \varphi(x,y,z,t)
\end{equation}
<br>

Fouriergesetzt: Wärmefluss proportional zum Gradienten der Temperatur. <br>
\begin{equation}
\vec{q} = - \lambda \nabla T
\end{equation}
<br>

Eingesetzt ergibt sich die Wärmeleitungsgleichung. Teilt man das Ergebnis durch $c \rho$, so ergibt sich die zweite Gleichung.<br>
\begin{equation}
\frac{d}{d t} (c \rho T) = \nabla \big( \lambda \nabla T \big) + \varphi(x,y,z,t)\\
\Leftrightarrow \quad \frac{d}{d t} T = \alpha \nabla^2 T + Q(x,y,z,t)
\end{equation}
<br>
mit $\alpha = \frac{\lambda}{c \rho}$
<br>

### Herleitung der Wärmeleitung über Integration

Gaußscher Integralsatz besagt, dass die Divergenz in einem Volumen gleich dem Fluss durch die Oberfläche ist.
<br>
\begin{equation}
\int_V \nabla (\vec{F}) dV = \oint_S \vec{F} \cdot \vec{n} dS
\end{equation}
<br>

Die Leistung (Energie/Zeit) für die Temperaturänderung = Leistung abgeführt durch die Wärmeflüsse und erzeugt durch die innnere Quellen.
<br>
\begin{equation}
\int_V \frac{\delta}{\delta t} (c \rho T) dV = \oint_S \vec{q} \cdot \vec{n} dS + \int_V \varphi(x,y,z,t) dV
\end{equation}
<br>

Darauf wird der Gaußsche Integralsatz angewendet, die Zeitableitung vors Integral gezogen und dann das Volumenintegral entfernt, weil es für alle möglichen Volumen $V$ gilt. Daraus ergibt sich die Wärmeleitungsgleichung.
<br>
\begin{equation}
\frac{d}{d t} (c \rho T) = \nabla \big( \vec{q}) + \varphi(x,y,z,t)
\end{equation}
<br>

Mit dem Fouriergesetzt $\vec{q} = - \lambda \nabla T$ ergibt sich die Wärmeleitungsgleichung:
<br>
\begin{equation}
\frac{d}{d t} (c \rho T) = \nabla \big( \lambda \nabla T \big) + \varphi(x,y,z,t)\\
\Leftrightarrow \quad \frac{d}{d t} T = \alpha \nabla^2 T + Q(x,y,z,t)
\end{equation}
<br>

### Randbedingungen

##### Dirichlet (konstante Temperatur)
$T(x_0,y_0,z_0) = T_0 = konstant$ <br>
##### Neumann (konstante Änderung/Wärmefluss)
$-\lambda \frac{d}{d n} T(x_0,y_0,z_0) = q(x_0,y_0,z_0) = konstant$ <br>
##### Newton-Richmann (Änderung/Wärmefluss proportional zum Temperaturunterschied)
$-\lambda \frac{d}{d n} T(x_0,y_0,z_0) = \alpha (T)(T-T_\infty)$ <br>
<br>

Newton-Richmann-Randbedingung wiederspricht der Wärmeleitungsgleichung, weil die Temperatur am Rand einen Knick hat(nicht differenzierbar). Diese Randbedingung ist in der Praxis wichtig, weil in einem Versuch die Wärmeleitungskoeffizienten ermittelt werden können und durch diese Randbedingung vereinfacht andere Effekte des Wärmetransports linear angenähert werden können, wie etwa Konvetion oder Wärmestrahlung.

### Formfunktionen

Temperatur an ausgewählten Nodes werden durch Formfunktionen auf das gesamte Gebiet extrapoliert. Hier z.B. lineare Interpolation:
<br>
\begin{equation}
\underbrace{T(x)}_{T}=
\underbrace{\begin{pmatrix}
1 & x
\end{pmatrix}}_{\overline{N}} \cdot
\underbrace{\begin{pmatrix}
\alpha_i\\
\alpha_j
\end{pmatrix}}_{\alpha}
\end{equation}
<br>

In unserem Beispiel mit linearen Interpolation können Temperaturen ($T_i,T_j$) an zwei Nodes ($x_i,x_j$) linear angepasst werden:
<br>
\begin{equation}
\underbrace{\begin{pmatrix}
T_i\\
T_j
\end{pmatrix}}_{a^e} =
\underbrace{\begin{pmatrix}
1 & x_i\\
1 & x_j
\end{pmatrix}}_{C} \cdot
\underbrace{\begin{pmatrix}
\alpha_i\\
\alpha_j
\end{pmatrix}}_{\alpha}
\end{equation}
<br>

Daraus können die Koeffizienten $\alpha = C^{-1} a^e$ berechnet werden, so dass die Temperatur T(x) als eine Matrixmultiplikation der Formfunktion $N^e = \overline{N} C^{-1}$ mit den Temperaturen an den Knoten $a^e$ über $T = N^e a^e$ dargestellt werden kann.
<br>
\begin{equation}
T =
\frac{1}{L}
\begin{pmatrix}
x_j & -x_i \\
-x & x
\end{pmatrix}
\begin{pmatrix}
T_i\\
T_j
\end{pmatrix}
\end{equation}
<br>

Dies ist äquivalent zu einer Darstellung mit elementaren Fomfunktionen $N^e_i$:
<br>
\begin{equation}
T = N^e_i T_i + N^e_j T_j
= \frac{1}{L}
\begin{pmatrix}
x_j-x & x-x_i
\end{pmatrix}
\begin{pmatrix}
T_i\\
T_j
\end{pmatrix}
\end{equation}
<br>

Diese können einfach mit der Lagrange-Polynomformel ausgedrückt werden:
<br>
\begin{equation}
T(x) = \sum_{k=1}^{n} l_k T_k \quad \textrm{mit} \quad l_k = \frac{(x-x_1)(x-x_2)...(x-x_{k-1})(x-x_{k+1})...(x_k-x_n)}{(x_k-x_1)(x-x_2)...(x_k-x_{k-1})(x_k-x_{k+1})...(x_k-x_n)}
\end{equation}
<br>

### schwache Formulierung

Um eine Differentialgleichung in die schwache Formulierung umzuformen, wird die DGL mit einer Gewichtsfunktion $v(x,y,z)$ multipliziert und dann über den Definitionsbereich integriert. Als Beispiel für die stationäre ($\frac{d}{dt} c \rho T(x,y,z)= 0$) Wärmeleitungsgleichung $\lambda \nabla^2 T(x,y,z) + \varphi(x,y,z,t) = 0$. Im unterem Beispiel ist die Temperatur $T(x)$ und erzeugte Wärme $\varphi(x)$ unabhängig von y,z,t, so dass über y,z integriert sich nur ein Faktor $A(x)$ ergibt und die Temperatur in einer x-Ebene gleich ist.
<br>
\begin{equation}
\int_V v(x) \cdot \left( \lambda \nabla^2 T(x) + \varphi(x) \right) d^3r = 0 \\
\Leftrightarrow \int_0^L v(x) \cdot A(x) \left( \lambda \nabla^2 T(x) + \varphi(x) \right) dx = 0 \\
\Leftrightarrow \int_0^L v(x) \lambda A(x) \nabla^2 T(x) dx + \int_0^L v(x) A(x) \varphi(x) dx = 0 \\
\Leftrightarrow \int_0^L v(x) \lambda A(x) \nabla^2 T(x) dx + \int_0^L v(x) A(x) \varphi(x) dx = 0
\end{equation}
<br>

Mit partieller Integration ergibt sich daraus die schwache Formulierung.(Im mehrdimensionalem Fall wird dazu das Green-Gauss-Theorem $\int_L \Phi \textbf{q}^T \textbf{n} dL = \int_A \Phi div(\textbf{q}) dA + \int_A (\nabla \Phi)^T \textbf{q} dA$ verwendet.) In der zweiten Gleichung wurde noch das Fouriergesetz $\vec{q} = - \lambda \nabla T$ angewandt.
<br>
\begin{equation}
\lambda A(x) v(x) (\nabla T(x)) \Big|_0^L - \int_0^L \lambda A(x) (\nabla v(x)) (\nabla T(x)) dx + \int_0^L v(x) A(x) \varphi(x) dx = 0\\
- A(x) v(x) q(x) \Big|_0^L - \int_0^L \lambda A(x) (\nabla v(x)) (\nabla T(x)) dx + \int_0^L v(x) A(x) \varphi(x) dx = 0
\end{equation}
In diese Gleichung können jetzt Randbedingungen für T,q eingesetzt werden<br>

Vorteile:<br>
- eine Ableitung aus der DGL auf die Gewichtsfunktion übertragen, welche leichter im Vorraus berechnet werden kann.
- kontinuierliche Übergänge im Lösungsbereich 0 bis L (z.B. bei Grenze unterschiedlicher Materialien mit unterschiedlichen Eigenschaften. Durch Zusammenfassung der Integrale in eins automatisch erfüllt (dann ist Materialeigenschaft z.B. $\alpha$ abschnittsweise definiert).
<br><br>


### Formulierung des linearen Gleichungssystems für FEM

Die gesuchte Lösung T(x) wird durch eine Linearkombination von Formfunktionen angenähert.<br> 
\begin{equation}
T(x) = \lbrack N_1(x) \; N_2(x) \; ... \; N_n(x) \rbrack \cdot \begin{bmatrix} T_1\\T_2\\...\\T_n \end{bmatrix}
\end{equation}

Als Gewichtsfunktionen kommen z.B. Deltafunktionen $v_i(x)=\delta(x_i-x)$ (point collocation), $v_i(x)=1$ im Intervall $\lbrack x_{i},x{i+1} \rbrack$ (subdomain collocation), Minimierung der Quadrate der Residuen an der Nullstelle des Gradienten => $v_i(x) = \frac{\partial e(x)}{\partial a_i}$ (least squares) oder die Formfunktionen $v_i(x)=N_i(x)$ (Galerkin) zum Einsatz. Die stationäre Wärmeleitungsgleichung $\lambda \nabla^2 T(x) + \varphi(x) = e(x) \approx 0$ wird durch die Näherung nicht exakt Null, sondern es bleibt ein Residuum $e(x)$ übrig. Die schwache Formulierung ergibt ein Integral über ein Produkt aus Gewichtsfunktion und stationärer Wärmeleitungsgleichung $\int_V v(x) \cdot \left( \lambda \nabla^2 T(x) + \varphi(x) \right) d^3r = 0$, welches auf Null gesetzt wird. Dies entspricht $\int_V v(x) e(x) d^3r = 0$, so dass durch diese Bedingung das Residuum klein gehalten wird.<br>
Im weiterem wird nur die Galerkin-Methode verwendet.
<br>
\begin{equation}
v(x) = \big( \lbrack N_1(x) c_1 \; N_2(x) c_2 \; ... \; N_n(x) c_n \rbrack \big) ^T
\end{equation}
Die Näherungen eingesetzt in die schwache Formulierung $\int_0^L \lambda A(x) (\nabla v(x)) (\nabla T(x)) dx = - A(x) v(x) q(x) \Big|_0^L + \int_0^L v(x) A(x) \varphi(x) dx$ und $c_i$ herausgezogen ergibt:<br>
\begin{equation}
\int_0^L \lambda A(x) \begin{bmatrix} \nabla N_1(x)\\ \nabla N_2(x)\\ ... \\ \nabla N_n(x) \end{bmatrix} \lbrack \nabla N_1(x) \; \nabla N_2(x) \; ... \; \nabla N_n(x) \rbrack \cdot \begin{bmatrix} T_1\\T_2\\...\\T_n \end{bmatrix} dx = - A(x) \begin{bmatrix} N_1(x)\\ N_2(x)\\ ... \\ N_n(x) \end{bmatrix} q(x) \Big|_0^L + \int_0^L A(x) \begin{bmatrix} N_1(x)\\ N_2(x)\\ ... \\ N_n(x) \end{bmatrix} \varphi(x) dx\\
\Leftrightarrow
\underbrace{\int_0^L \lambda A(x) \begin{bmatrix} \nabla N_1(x) \nabla N_1(x) &  \nabla N_1(x)  \nabla N_2(x) & ... &  \nabla N_1(x) \nabla N_n(x)\\ \nabla N_2(x) \nabla N_1(x) &  \nabla N_2(x)  \nabla N_2(x) & ... &  \nabla N_2(x) \nabla N_n(x)\\ ...&...&...&...\\ \nabla N_n(x) \nabla N_1(x) &  \nabla N_n(x)  \nabla N_2(x) & ... &  \nabla N_n(x) \nabla N_n(x)\\ \end{bmatrix} dx}_{\text{stiffness matrix }K} \cdot \underbrace{\begin{bmatrix} T_1\\T_2\\...\\T_n \end{bmatrix}}_{\text{Temperatures } a} = - \underbrace{A(x) \begin{bmatrix} N_1(x)\\ N_2(x)\\ ... \\ N_n(x) \end{bmatrix} q(x) \Big|_0^L}_{\text{boundary vector } f_b} + \underbrace{\int_0^L A(x) \begin{bmatrix} N_1(x)\\ N_2(x)\\ ... \\ N_n(x) \end{bmatrix} \varphi(x) dx}_{\text{load vector } f_l}
\end{equation}
<br>

### nicht-stationärer Fall
Die schwache Formulierung der DGL kann im nicht-stationären Fall analog durchgeführt werden. Das Ergebnis ist:<br>
\begin{equation}
\underbrace{\int_0^L v(x) \rho c \frac{\partial a)}{\partial t} dx}_{M \frac{\partial T_i}{\partial t}} + \underbrace{\int_0^L \lambda A(x) (\nabla v(x)) (\nabla T(x,t)) dx}_{K a} = - \underbrace{A(x) v(x) q(x,t) \Big|_0^L}_{f_b} + \underbrace{\int_0^L v(x) A(x) \varphi(x,t) dx}_{f_l}
\end{equation}
Dabei bekommt man ein Gleichungssystem, welches nach der Zeitableitung umgestellt wird.<br>
\begin{equation}
\frac{\partial a}{\partial t} = M^{-1} \big( -Ka + f_b + f_l \big)
\end{equation}
Dabei ist $a = \lbrack T_1 \; T_2 \; ... \; T_n \rbrack ^T$<br>
Um diese Gleichung zu lösen, muss ein Anfangswert für $a$ angegeben werden und dann die Lösung mit einem numerischen Verfahren in kleinen Zeitschritten entwickelt werden. Das einfachste Verfahren dazu ist das Eulerverfahren.<br>
\begin{equation}
a(t+\Delta t) = a(t) + \Delta t \cdot \frac{\partial a(t)}{\partial t}
\end{equation}

## Formulierung eines 1D Beispiel-Problems

Querschnittsfläche $A = 10 m^2$<br>
Wärmeleitfähigkeit $\lambda = 5 \frac{W}{m K}$<br>
Wärmequelle-Dichte $\varphi = 10 \frac{W}{m^3}$<br>

Knoten $x \in\{2m,4m,6m,8m\}$<br>

Randbedingungen:<br>
Temperatur: $T(2m)=0°C$<br>
Wärmefluss: $q(8m)=15 \frac{W}{m^2}$<br>

## analytische Lösung
exakt:

\begin{equation}
\underbrace{\frac{d}{d t} (c \rho T)}_{=0} = \nabla \big( \lambda \nabla T \big) + \varphi(x,y,z,t)\\
\Leftrightarrow -\lambda \nabla^2 T = \varphi = 10 \frac{W}{m^3}
\end{equation}

integrieren:
\begin{equation}
-\lambda \nabla T = 10 \frac{W}{m^2} \cdot x + C_1
\end{equation}

Fouriergesetzt:
\begin{equation}
\vec{q} = - \lambda \nabla T
\end{equation}

angewendet:
\begin{equation}
q = 10 \frac{W}{m^3} \cdot x + C_1
\end{equation}

mit Randbedingung:
\begin{equation}
q(8m) = 80 \frac{W}{m^2} + C_1 \stackrel{!}{=} 15 \frac{W}{m^3} \\
\Rightarrow C_1 = -65 \frac{W}{m^2} \\
\Rightarrow q = - \lambda \nabla T = 10 \frac{W}{m^3} \cdot x -65 \frac{W}{m^2}
\end{equation}

zurück umgeschrieben:
\begin{equation}
q = - \lambda \nabla T = 10 \frac{W}{m^3} \cdot x -65 \frac{W}{m^2} \\
\nabla T = -2 \frac{K}{m^2} \cdot x +13 \frac{K}{m}
\end{equation}

und nochmal integriert:
\begin{equation}
T = - 1 \frac{K}{m^2} \cdot x^2 +13 \frac{K}{m} \cdot x + C_2
\end{equation}

mit Randbedingung:
\begin{equation}
T(2m) = - 4°C + 26°C + C_2 \stackrel{!}{=} 0°C \\
\Rightarrow C_2 = - 22 °C
\end{equation}

folg das Ergebnis zu:
\begin{equation}
\Rightarrow  T = - 1 \frac{K}{m^2} \cdot x^2 +13 \frac{K}{m} \cdot x - 22 °C
\end{equation}

```python
#Nodes
class NodePoints:
    def __init__(self, node_id=None, position=None):
        self.__node_id = node_id
        self.__node_position = position
##################################################
    def get_node_position(self):
        return self.__node_position
    def get_node_id(self):
        return self.__node_id
    def set_node_position(self,node_position):
        self.__node_position = node_position
        return None
    def set_node_id(self,node_id):
        self.__node_id = node_id
        return None
##################################################
```

```python
#Elements
class Elements:
    def __init__(self, fem_id=None, node_names=[]):
        self.__fem_id = fem_id
        self.__fem_nodes = node_names
##################################################
    def get_fem_id(self):
        return self.__fem_id
    def get_fem_nodes(self):
        return self.__fem_nodes
    def set_fem_id(self,fem_id):
        self.__fem_id = fem_id
        return None
    def set_fem_nodes(self):
        self.__fem_nodes = fem_nodes
        return None
##################################################
    def add_node(self,node_name):
        self.__fem_nodes.append(node_name)
        return None
    
    def fem_node_position(self,node_nr):
        return self.__fem_nodes[node_nr].get_node_position()
    
    def node_distance(self,node0,node1):
        node0_position = self.fem_node_position(node0)
        node1_position = self.fem_node_position(node1)
        return node1_position - node0_position
    
    def local_form_function(self):
        L = self.node_distance(0,1)
        f0 = lambda x: (x - self.fem_node_position(0))/L
        f1 = lambda x: (self.fem_node_position(1) - x)/L
        return f0,f1
    
    def derivative_local_form_function(self):
        L = self.node_distance(0,1)
        f0 = lambda x: 1/L
        f1 = lambda x: -1/L
        return f0,f1
    
    def plot_local_form_function(self,f0,f1):
        plot_range = [self.fem_node_position(0),self.fem_node_position(1)]
        x = np.linspace(min(plot_range),max(plot_range),200)
        fig = plt.figure(num=None, figsize=(10,5), dpi=80, facecolor='w', edgecolor='k')
        ax0=plt.subplot() #2x1 subplots, select 1 plot
        ax0.scatter(x,f0(x),color="green",label="function", linewidth=2, s=2, zorder=2)
        ax0.scatter(x,f1(x),color="blue",label="function", linewidth=2, s=2, zorder=2)
        ax0.axhline(0, color='black', lw=1)
        ax0.legend(fontsize=12, bbox_to_anchor=(0.9,0.9))
        ax0.set_title('Function', fontsize=16)
        ax0.set_xlabel('X', fontsize=12)
        ax0.set_ylabel('Y', fontsize=12)
        plt.show()
        return None
    def local_stiffness_matrix(self,A,k):
        L = self.node_distance(0,1)
        return np.array([[1, -1], [-1, 1]])/L*A*k
```

```python
class EQsystem:
    def __init__(self, *args, A=None, k=None, phi=None, q0=None, qn=None, T0=None, Tn=None):
        self.__elements = []
        self.__K = None;
        self.__fl = None;
        self.__fb = None;
        self.__f = None;
        self.__T = None;
        ###constants
        self.__A = A;
        self.__k = k;
        self.__phi = phi;
        self.__q0 = q0;
        self.__qn = qn;
        self.__T0 = T0;
        self.__Tn = Tn;
        for x in args:
            self.__elements.append(x)
  #      for arg in kwargs:
   #         arg;
    #        arg.value()
##################################################
    def append_fem(self,*args):
        for x in args:
            self.__elements.append(x)
        return None
##################################################
    def global_stiffness_matrix(self):
        n = len(self.__elements)
        self.__K = np.zeros((n+1,n+1))
        for iii in range(0,n):
            element = self.__elements[iii]
            self.__K[iii:iii+2,iii:iii+2] = self.__K[iii:iii+2,iii:iii+2] + element.local_stiffness_matrix(self.__A,self.__k)
        return self.__K
    def boundary_vector(self):
        n = len(self.__elements)
        self.__fl = np.zeros(n+1)
        self.__fl[0]=self.__q0;self.__fl[-1]=self.__qn;
        self.__fl = - self.__fl*self.__A
        return self.__fl
    def load_vector(self):
        n = len(self.__elements)
        self.__fb = - np.zeros(n+1)
        for jjj in range(0,n+1): #0,1,2,3
            try:
                if (jjj-1>=n) or (jjj-1>=0):
                    element = self.__elements[jjj-1]#(-1),0,1,2
                    self.__fb[jjj] += element.node_distance(0,1)/2
            except:
                pass
            try:
                if (jjj<=n) or (jjj>=0):
                    element = self.__elements[jjj]#0,1,2,(3)
                    self.__fb[jjj] += element.node_distance(0,1)/2
            except:
                continue
        self.__fb = self.__fb*self.__A*self.__phi
        return self.__fb
    def solution(self):
        n = len(self.__elements)
        self.__T = np.zeros(n+1)
        
        self.global_stiffness_matrix()
        self.boundary_vector()
        self.load_vector()
        
        self.__f = self.__fl + self.__fb;
        
        T_offset=0
        if self.__q0 == None:
            self.__K = self.__K[1:,1:]
            self.__f = self.__f[1:]
            T_offset = 1
        if self.__qn == None:
            self.__K = self.__K[:-1,:-1]
            self.__f = self.__f[:-1]
        
        print('\n',self.__K,'\n')
        print('\n',self.__f,'\n')
        
        T_solution = np.linalg.solve(self.__K,self.__f)
        
        if self.__T0 != None:
            self.__T[1:] = T_solution
            self.__T[0] = self.__T0;
        elif self.__Tn != None:
            self.__T[0:-1] = T_solution
            self.__T[-1] = self.__Tn;
        else:
            self.__T = T_solution
        return self.__T
    def plot_solution(self):
        fig = plt.figure(num=None, figsize=(10,5), dpi=80, facecolor='w', edgecolor='k')
        ax0=plt.subplot() #2x1 subplots, select 1 plot
        
        x=np.linspace(2,8,101)
        ax0.plot(x,(-x**2+13*x-22),color="green",label="analytic", linewidth=2, zorder=2)
        
        n = len(self.__elements)
        postions = set()
        for element in self.__elements:
            postions.add(element.fem_node_position(0))
            postions.add(element.fem_node_position(1))
        postions = list(postions)
        postions.sort()
        ax0.scatter(postions,self.__T,color="blue",label="fem", linewidth=2, s=8, zorder=2)
        
        ax0.axhline(0, color='black', lw=1)
        ax0.legend(fontsize=12, bbox_to_anchor=(0.9,0.9))
        ax0.set_title('Temperature', fontsize=16)
        ax0.set_xlabel('Position', fontsize=12)
        ax0.set_ylabel('Temperature', fontsize=12)
        plt.show()
        return None
```
